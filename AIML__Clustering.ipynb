```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering, KMeans
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate random data
np.random.seed(42)
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Hierarchical clustering
linkage_matrix = linkage(X, method='ward')
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Cut the dendrogram to get clusters
num_clusters = 4
hierarchical_labels = AgglomerativeClustering(n_clusters=num_clusters, linkage='ward').fit_predict(X)

# KMeans clustering
kmeans_labels = KMeans(n_clusters=num_clusters, random_state=42).fit_predict(X)

# Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=hierarchical_labels, cmap='viridis')
plt.title('Hierarchical Clustering')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('KMeans Clustering')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, KMeans

# Generate non-spherical data
X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

# Apply DBSCAN
dbscan_labels = DBSCAN(eps=0.3, min_samples=5).fit_predict(X)

# Apply KMeans
kmeans_labels = KMeans(n_clusters=2, random_state=42).fit_predict(X)

# Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('KMeans Clustering')

plt.show()
```
