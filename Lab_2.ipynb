# 1)Change the coefficients array and try a different polynomial instead of our x2

# 3)Take a degree 5 polynomial with 5 roots and try different initializations, instead of random ones. Does it converge on different values for different initializations? Why does initialization not matter in our case of x2?

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients for the degree 5 polynomial
coefficients = [1, -5, 10, -10, 5, -1]

# Generate x values
x = np.linspace(-2, 3, 100)

# Calculate the corresponding y values using the degree 5 polynomial
y = np.polyval(coefficients, x)

# Plot the degree 5 polynomial
plt.plot(x, y, label='Degree 5 Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Degree 5 Polynomial')
plt.grid(True)
plt.show()

# Find the roots of the polynomial
roots = np.roots(coefficients)

print("Roots of the polynomial:", roots)
```

# 4)Can you modify the algorithm to find a maxima of a function, instead of a minima?

```python
import numpy as np
import matplotlib.pyplot as plt

def initialize_parameters():
    # Replace this with your actual parameter initialization logic
    # For example, initialize theta to a starting value
    return 0.0

def compute_gradient(theta):
    # Replace this with your actual gradient computation logic
    # This is the gradient of your loss function with respect to the parameters
    # For simplicity, let's assume a simple quadratic loss for illustration
    return -2 * theta  # Negative gradient for gradient ascent

def compute_loss(theta):
    # Replace this with your actual loss computation logic
    # This is the objective function you want to maximize
    # For simplicity, let's assume a simple quadratic loss for illustration
    return -theta**2  # Negative loss for gradient ascent

def convergence_criteria_met(loss):
    # Replace this with your actual convergence criteria
    # For simplicity, let's assume convergence when the loss is below a threshold
    return loss < 1e-5

def gradient_ascent(learning_rate, max_iterations):
    # Initialize parameters or weights
    theta = initialize_parameters()

    for iteration in range(max_iterations):
        # Compute the gradient of the loss with respect to the parameters
        gradient = compute_gradient(theta)

        # Update the parameters using the learning rate and the gradient
        theta = theta + learning_rate * gradient

        # Compute the loss
        loss = compute_loss(theta)

        # Check for convergence (you may use a convergence threshold)
        if convergence_criteria_met(loss):
            print(f"Converged in {iteration} iterations. Final loss: {loss}")
            break

    return theta

# Example usage with different learning rates
learning_rates = [0.1, 0.01, 0.001, 0.0001]

for lr in learning_rates:
    print(f"\nTraining with learning rate: {lr}")
    theta_optimal = gradient_ascent(learning_rate=lr, max_iterations=1000)

# Plot the function and the optimal point found by gradient ascent
x = np.linspace(-10, 10, 100)
y = -x**2  # Objective function

plt.plot(x, y, label='Objective Function')
plt.scatter(theta_optimal, -theta_optimal**2, color='red', label='Optimal Point')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Gradient Ascent for Maximizing a Quadratic Function')
plt.grid(True)
plt.show()
```
